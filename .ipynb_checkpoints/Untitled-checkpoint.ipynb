```python
import pickle as pkl
import matplotlib.pyplot as plt
import numpy as np
```

```python
distances = np.load("distances.pkl", allow_pickle=True)
dimer_energies = np.load("dimer_energies.pkl", allow_pickle=True)
charges_energies = np.load("charge_coupling_energies.pkl", allow_pickle=True)
dipole_energies = np.load("dipole_coupling_energies.pkl", allow_pickle=True)
orient_energies = np.load("orient_coupling_energies.pkl", allow_pickle=True)
```

```python
fig, ax = plt.subplots()

ax.scatter(distances, dimer_energies)
ax.scatter(distances, charges_energies)
ax.scatter(distances, dipole_energies)
ax.scatter(distances, orient_energies)
ax.set_ylim(0.09, 0.1)
ax.set_xlim(0.09, 0.1)

fig.set_size_inches(12,12)
```

```python
def two_by_two_mat(v):
    mat = np.array([0, v, v, 1])
    mat = mat.reshape((2,2))

    vals, vecs = np.linalg.eig(mat)
    
    return vals

higher = []
lower = []

v = np.linspace(-1, 1, 1000)

for i in v:
    vals = two_by_two_mat(i)
    higher.append(max(vals))
    lower.append(min(vals))
    
fig, ax = plt.subplots()

ax.plot(v, higher)
ax.plot(v, lower)
```
